```python
def create_overlay_plot(start_date, end_date=None, **plt_properties):
    """
    A helper method to draw overlay plots between Bitstamp and Coinbase.
    Calling this method with the appropriate variable will plot the  charts.
    :param start_date: Starting date of time period for data in axis
    :param end_date: Ending date of time period for data in axis
    :param plt_properties: kwargs with fig_size, title, bitstamp_color, coinbase_color
    :return: None
    """
    if end_date is None:
        # If end_date is not provided, it is the assumption that the user wants to plot
        # data for a single day. Hence, the end_date is set to start_date.
        end_date = start_date

    # Keyword arguments which are optional with default value set.
    # If the arguments are not provided, default values set here are used.
    fig_size = plt_properties.get('fig_size',(30,15))
    title = plt_properties.get('title', 'Line Plot with Pandas')
    bitstamp_color = plt_properties.get('bitstamp_color','orchid')
    coinbase_color = plt_properties.get('coinbase_color','turquoise')

    # Plot first axes with bitstamp data for the given start date and end date.
    ax = bitstamp_sliced.loc[start_date:end_date].plot(figsize=fig_size, color=bitstamp_color, label='Bitstamp')
    # Plot second axes with coinbase data for the given start data and end date.
    # Overlay the bitstamp axes on the coinbase axes with `ax=ax` option.
    coinbase_sliced.loc[start_date:end_date].plot(figsize=fig_size, title=title, color=coinbase_color, label='Coinbase', ax=ax)
    # Add a legend
    ax.legend(['Bitstamp','Coinbase'])
    
```

```python
# Overlay the visualizations for the bitstamp and coinbase DataFrames in one plot
# The plot should visualize the prices over the full lenth of the dataset
# Be sure to include the parameters: legend, figure size, title, and color and label

create_overlay_plot('2018-01-01', '2018-03-31', fig_size=(30,10), title="Bitcoin: Bitstamp vs Coinbase Closing Price")
```

#### 2. For each of the three dates, calculate the spread returns. To do so, divide the instances that have a positive arbitrage spread (that is, a spread greater than zero) by the price of Bitcoin from the exchange you’re buying on (that is, the lower-priced exchange). Review the resulting DataFrame.

```python
def calc_spread_return(date, arbitrage_spread, lower_exchange):
    """
    Caluclate spread returns for the given date, lower exchange and the arbitrage spread data frame
    :param date: Date for which the spread returns have to be calculated.
    :param arbitrage_spread: Arbitrage spread data frame with spread data between bitstamp and coinbase.
    :param lower_exchange: Exchange name with the lower prices.
    :return: Return spread data frame
    """
    return arbitrage_spread[arbitrage_spread['Close'] > 0] / lower_exchange.loc[date]
```

#### 5. For each of the three dates, calculate the potential profit, in dollars, per trade. To do so, multiply the spread returns that were greater than 1% by the cost of what was purchased. Make sure to drop any missing values from the resulting DataFrame.

```python
def calc_potential_profits(profitable_trades, date, exchage_bought):
    """
    Calculate the potential profit, in dollars, per trade.
    Potential Profit = (Spread Returns > 0.01) * Purchase Price
    :param profitable_trades: Data frame with Spread Returns > 0.01 for the given date
    :param date: Date for which potential profits have to be calculated.
    :param exchage_bought: Exchange used for purchasing bitcoin.
    :return: Data frame with profits.
    """
    profit = profitable_trades * exchage_bought.loc[date]
    profit.dropna(inplace=True)
    return profit
```

```python
# Calculate potential profits for '2018-03-26'
profit_per_trade_late = calc_potential_profits(profitable_trades_late, '2018-03-26', bitstamp_sliced)
profit_per_trade_late
```
